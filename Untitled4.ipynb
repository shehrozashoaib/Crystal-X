```python
from dataclasses import fields
from atom2vec import AtomSimilarity
from pymatgen.core import Structure,Element
from typing import List

element_symbols = [element.symbol for element in Element]
print(element_symbols)





results = mpr.materials.search(fields=["structure"])


#results = mpr.summary.search(elements=["Si","O"],fields = ["structure"])
#print(results)
```

```python
print(results[0].structure)
```

```python
entries = [entry.structure for entry in results]
```

```python
atom_similarity = AtomSimilarity.from_structures(entries,
                                                 k_dim=50, max_elements=3)
```

```python
from pymatgen.core import Element
```

```python
calcium = Element("Ca")
strontium = Element("Sr")


print(atom_similarity.get_atom_similarity((calcium,strontium)))
print(atom_similarity.get_atom_similarity((Element("H"),(Element("H")))))
```

```python
element_symbols = [element.symbol for element in Element]

# Print all element symbols
print("List of all elements:")
for symbol in element_symbols:
    print(symbol)
```

```python
import numpy as np
```

```python
val_array = np.zeros((len(element_symbols),len(element_symbols))).astype(float)
val_array
```

```python
for elem in element_symbols:
    for elem2 in element_symbols:
        val = float(atom_similarity.get_atom_similarity((Element(elem),Element(elem2))))


        val_array[element_symbols.index(elem),element_symbols.index(elem2)] = val
        print(val_array[element_symbols.index(elem),element_symbols.index(elem2)])

print(val_array)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
cmap = mcolors.LinearSegmentedColormap.from_list(
    'custom_cmap', [(0, 'darkblue'), (0.5, 'yellow'), (1, 'darkred')]
)

# Plot the heatmap
plt.imshow(val_array, cmap=cmap, vmin=-1, vmax=1)
plt.colorbar(label='Value')
plt.title('Heatmap of Array')
plt.show()
```
